# 自定義模型(Model)
### 演示單一神經元計算過程


## 定義計算過程

```python
class Model(object):
    def __init__(self):
        self.w = tf.Variable(5.0) #初始權重(weight)為5.0
        self.b = tf.Variable(0.0) #初始偏差(bias)為5.0
        
    def __call__(self,x):
        result = self.w*x + self.b
        return result
```

## 呼叫與執行

```python
input_data = 3.0 # 可自行調整值看看

model = Model()
print('當輸入為:' + str(input_data) + ' 模型輸出為:' + str(model(input_data).numpy()))
```

## 定義損失函數(loss function)

```python
def loss(predicted_y, desired_y):
    now_loss = tf.reduce_mean(tf.square(predicted_y - desired_y))
    return now_loss
```

## 隨機生成數據

```python
TRUE_w = 3.0
TRUE_b = 2.0
Num_examples = 1000

inputs = tf.random.normal(shape=[Num_examples])
noise = tf.random.normal(shape=[Num_examples])

outputs = inputs*TRUE_w + TRUE_b + noise
```

## 數據可視化(訓練前)
## 當前loss
<font color = blue size = 4> 藍色為生成的數據 </font>

<font color = red size = 4> 紅色為模型的預測 </font>

```python
plt.scatter(inputs, outputs, c='b')

predict = model(inputs) # 輸入模型
plt.scatter(inputs, predict, c='r')
plt.show() # 即時顯示

print('loss:' + str(loss(predict,outputs).numpy())) # 顯示loss
```

### 定義訓練函數

```python
def train(model, inputs, outputs, learning_rate):
    with tf.GradientTape() as t:
        current_loss = loss(model(inputs), outputs)
    dw, db = t.gradient(current_loss, [model.w, model.b])
    model.w.assign_sub(learning_rate*dw)
    model.b.assign_sub(learning_rate*db)
```

# 開始訓練~~

```python
model = Model()

# 收集loss、權重與偏差的歷史值
ws = []
bs = []
loss_history = []
# 定義訓練次數
epochs = 15 # 可自行調整
learning_rate = 0.1 # 可自行調整
# 訓練network
for epoch in range(epochs):
    ws.append(model.w.numpy())
    bs.append(model.b.numpy())
    loss_history.append(loss(model(inputs), outputs))
    
    train(model, inputs, outputs, learning_rate)
    
    print('Now epoch : ' + str(epoch + 1) + ' ,Now weight : ' + str(ws[-1]) + ' ,Now bias : ' + str(bs[-1])
          + ' ,Now loss : ' + str(loss_history[-1].numpy()))

print('訓練結束')
```
